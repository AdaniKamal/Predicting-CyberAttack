# Analyse - Key Featured

```python
import os
import pandas as pd
import numpy as np

base_dir = os.path.expanduser("~/Documents/FYP_Project")
data_dir = os.path.join(base_dir, "data")
processed_dir = os.path.join(data_dir, "processed")

df_kaggle = pd.read_csv(os.path.join(processed_dir, "kaggle_clean.csv"))
df_nvd = pd.read_csv(os.path.join(processed_dir, "nvd_clean.csv"))
df_exploitdb = pd.read_csv(os.path.join(processed_dir, "exploitdb_clean.csv"))
df_tenable = pd.read_csv(os.path.join(processed_dir, "tenable_clean.csv"))

df_kaggle.shape, df_nvd.shape, df_exploitdb.shape, df_tenable.shape
```

```python
# Enrich the Tenable data with real exploit --> ExploitDB
df_merge = df_tenable.merge(
    df_exploitdb[["cve_id", "verified_flag"]],
    on="cve_id",
    how="left"
)

df_merge["verified_flag"] = df_merge["verified_flag"].fillna(0)
df_merge.shape
# verified_flag = 1  public exploit exists
# verified_flag = 0  no known exploit
```

```python
# Merge vulnerability descriptions (NVD)
df_merge = df_merge.merge(
    df_nvd[["cve_id", "description"]],
    on="cve_id",
    how="left",
    suffixes=("", "_nvd")
)

df_merge.shape
```

## Create ATTACK TYPE labeling function (RF)

```python
def label_attack_type(row):
    text = " ".join([
        str(row.get("title", "")),
        str(row.get("description", "")),
        str(row.get("family", "")),
    ]).lower()

    exploit = row.get("verified_flag", 0)

    if "remote code execution" in text or "rce" in text or "buffer overflow" in text:
        return "RCE"
    if "privilege escalation" in text or "elevation of privilege" in text:
        return "PRIV_ESC"
    if "sql" in text or "injection" in text or "command execution" in text:
        return "INJECTION"
    if (
        "authentication bypass" in text
        or "authorization bypass" in text
        or "idor" in text
        or "insecure direct object reference" in text
    ):
        return "AUTH_BYPASS"
    if "cross-site scripting" in text or "xss" in text:
        return "XSS"
    if "path traversal" in text or "directory traversal" in text:
        return "TRAVERSAL"
    if "ssrf" in text or "server-side request forgery" in text:
        return "SSRF"
    if "denial of service" in text or "dos" in text or "ddos" in text:
        return "DOS"
    if "information disclosure" in text or "information leak" in text:
        return "INFO_DISC"
    if "misconfiguration" in text:
        return "MISCONFIG"

    # fallback logic
    if exploit == 1:
        return "RCE"

    return "MISCONFIG"
```

```python
df_merge["attack_type"] = df_merge.apply(label_attack_type, axis=1)
df_merge["attack_type"].value_counts()
```

## Add BANK RELEVANCE

```python
BANK_CORE_KEYWORDS = [
    # Banking domain words
    "bank", "banking", "core banking", "retail banking",
    "internet banking", "mobile banking", "digital banking",
    "financial institution", "finance",

    # Payments / cards / transfers
    "payment", "payments", "payment gateway", "payment switch",
    "credit card", "debit card", "cardholder", "pos", "point of sale",
    "atm", "swift", "sepa", "iso 20022", "remittance", "wire transfer",
    "merchant", "acquiring", "issuing",

    # AML/KYC/Fraud (bank-specific)
    "aml", "anti-money laundering", "kyc", "know your customer",
    "fraud", "fraudulent", "sanctions screening"
]

BANK_BANKING_TECH_VENDORS = [
    # Common banking/core systems vendors/products (not exhaustive)
    "temenos", "t24", "transact",
    "fis", "fiserve", "finacle",
    "oracle flexcube", "flexcube",
    "sap banking", "sopra", "mambu"
]

BANK_INFRA_KEYWORDS = [
    # Tech that is very common in banks (supporting infra)
    "active directory", "ldap", "kerberos",
    "vpn", "proxy", "waf", "reverse proxy",
    "hsm", "pkcs", "tls", "ssl",
    "siem", "splunk",
    "kubernetes", "docker",
    "citrix", "vmware", "windows server", "rhel", "linux"
]

# Things that often indicate non-bank specific consumer/creator products
LOW_RELEVANCE_HINTS = [
    "game", "gaming", "minecraft", "steam", "xbox",
    "wordpress theme", "photo editor", "music player"
]
```

```python
def label_bank_relevance_domain(row):
    text = " ".join([
        str(row.get("title", "")),
        str(row.get("description", "")),
        str(row.get("family", "")),
    ]).lower()

    score = 0

    # Strong signals: core banking & payment words
    if any(k in text for k in BANK_CORE_KEYWORDS):
        score += 3

    # Strong signals: banking vendors/products
    if any(k in text for k in BANK_BANKING_TECH_VENDORS):
        score += 4

    # Medium signals: banking-like enterprise infra
    if any(k in text for k in BANK_INFRA_KEYWORDS):
        score += 1

    # Reduce score if clearly consumer/non-enterprise context
    if any(k in text for k in LOW_RELEVANCE_HINTS):
        score -= 2

    # Convert score  label
    # You can tune these thresholds after you see distribution
    if score >= 4:
        return "HIGH"     # very likely bank/financial relevant
    elif score >= 1:
        return "MEDIUM"   # likely relevant to bank environments (infra)
    else:
        return "LOW"      # generic / not specifically bank-relevant
```

```python
df_merge["bank_relevance"] = df_merge.apply(label_bank_relevance_domain, axis=1)
df_merge["bank_relevance"].value_counts()
```

# FEATURE ENGINEERING

```python
# Prepare working dataframe
df_ml = df_merge.copy()

df_ml = df_ml[
    [
        "cvss_score",
        "severity",
        "family",
        "verified_flag",
        "bank_relevance",
        "attack_type"
    ]
]

df_ml.shape
```

## Handle MISSING values

```python
# Handle MISSING values
# CVSS: fill with median
df_ml["cvss_score"] = df_ml["cvss_score"].fillna(df_ml["cvss_score"].median())

# Categorical columns: fill unknown
for col in ["severity", "family", "bank_relevance"]:
    df_ml[col] = df_ml[col].fillna("UNKNOWN")
```

## Encode categorical features

```python
# Encode categorical features
from sklearn.preprocessing import LabelEncoder

encoders = {}

for col in ["severity", "family", "bank_relevance", "attack_type"]:
    le = LabelEncoder()
    df_ml[col] = le.fit_transform(df_ml[col].astype(str))
    encoders[col] = le
```

```python
# Define RF dataset (X, y)
X = df_ml.drop(columns=["attack_type"])
y_rf = df_ml["attack_type"]

X.shape, y_rf.shape
```

## Train/Test Split (70/30)

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train_rf, y_test_rf = train_test_split(
    X,
    y_rf,
    test_size=0.30,      # 30% test
    random_state=42,
    stratify=y_rf
)

X_train.shape, X_test.shape
```

## Train the model - RF

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=300,        # more trees = stabler probabilities for Top-5
    min_samples_split=5,
    min_samples_leaf=2,
    class_weight="balanced",
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train_rf)
```

## Result

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred_rf = rf.predict(X_test)

acc = accuracy_score(y_test_rf, y_pred_rf)
acc
```

```python
print(classification_report(y_test_rf, y_pred_rf))
```

```python
# Get probabilities
import numpy as np

y_proba = rf.predict_proba(X_test)
y_proba.shape
```

```python
# Decode labels back to attack names
attack_encoder = encoders["attack_type"]
attack_labels = attack_encoder.classes_

attack_labels
```

```python
# Top-5 prediction function + test one row
def top_k_attacks(prob_row, labels, k=5):
    idx = np.argsort(prob_row)[::-1][:k]
    return [(labels[i], float(prob_row[i])) for i in idx]

# Test on one sample
top_k_attacks(y_proba[0], attack_labels, k=5)
```

```python
def top_k_accuracy(y_true, y_proba, k=5):
    y_true_arr = np.array(y_true)
    correct = 0
    for i in range(len(y_true_arr)):
        topk_idx = np.argsort(y_proba[i])[::-1][:k]
        if y_true_arr[i] in topk_idx:
            correct += 1
    return correct / len(y_true_arr)

top5_acc = top_k_accuracy(y_test_rf, y_proba, k=5)
top5_acc
```
